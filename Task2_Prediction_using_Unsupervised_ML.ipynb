this task is cluster the iris dataset, and predict the number of clusters. 


```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

iris = datasets.load_iris()
print(iris)
```

```python
X = iris.data
y = iris.target
names = iris.target_names
print(X , y)
```

Vis. dataset in 2D and 3D before clustering. But dataset has 2 or 3 clusters.

```python
X_2d = PCA(n_components= 2).fit_transform(X,y)
plt.scatter(X_2d[:,0] , X_2d[:,1] )
X_3d = PCA(n_components = 3).fit_transform(X,y)
fig = plt.figure(figsize=(10, 10))
ax = Axes3D(fig)
ax.scatter(X_3d[:,0] , X_3d[:,1], X_3d[:,2])
plt.show()
```

Now, we try to answer **"How many clusters?"**. So, we need to meaure to select k. We propose two ways(i.e., two measures). first, model.inertia_ is Sum of squared distances of samples to their closest cluster center.  




```python
import random
measure = []
Ks = range(2,10)
for k in Ks:
  color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(k)]
  model = KMeans(n_clusters= k).fit(X)
  lables = model.labels_
  lables = [color[i] for i in lables]
  plt.scatter(X_2d[:,0] , X_2d[:,1], c = lables)
  fig = plt.figure(figsize=(10, 10))
  ax = Axes3D(fig)
  ax.scatter(X_3d[:,0] , X_3d[:,1], X_3d[:,2] , c = lables)
  plt.show()
  measure.append(model.inertia_)

plt.plot(Ks, measure)
plt.show()
```

Second, v_measure_score measures the qualtiy of clustring given true labels.



```python
from sklearn.metrics import v_measure_score
import random

measure = []
Ks = range(2,10)
for k in Ks:
  color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(k)]
  model = KMeans(n_clusters= k).fit(X)
  lables = model.labels_
  lables = [color[i] for i in lables]
  plt.scatter(X_2d[:,0] , X_2d[:,1], c = lables)
  fig = plt.figure(figsize=(10, 10))
  ax = Axes3D(fig)
  ax.scatter(X_3d[:,0] , X_3d[:,1], X_3d[:,2] , c = lables)
  plt.show()
  measure.append(v_measure_score(lables , y))

plt.plot(Ks, measure)
plt.show()
```

```python
import random
color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(3)]
model = KMeans(n_clusters= 3).fit(X)
lables = model.labels_
lables = [color[i] for i in lables]
plt.scatter(X_2d[:,0] , X_2d[:,1], c = lables)
fig = plt.figure(figsize=(10, 10))
ax = Axes3D(fig)
ax.scatter(X_3d[:,0] , X_3d[:,1], X_3d[:,2] , c = lables)
plt.show()
```
